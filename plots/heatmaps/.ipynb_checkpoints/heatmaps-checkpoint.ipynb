```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
sns.set_context('paper', font_scale=1.3)
```

```python
df_MAG_cellulosomes = pd.read_csv("MAGs_cellulosomes_DF.csv", index_col=0)
df_MAG_cellulosomes.head()
```

```python
df_MAG_cellulases = pd.read_csv("MAGs_cellulases_DF.csv", index_col=0)
df_MAG_cellulases.head()
```

```python
df_MAG_abundance = pd.read_csv("MAGs_abundance.csv", index_col=0)
df_MAG_abundance['sum'] = df_MAG_abundance.sum(axis=1)
df_MAG_abundance.head()
```

```python
mpl.rcParams['figure.figsize'] = (8,14)
```

```python
heatmap_seaborn = sns.heatmap(df_MAG_cellulases, vmin=0, vmax=10, cmap=mpl.cm.Greens, cbar_kws={'orientation':'horizontal'})
locs, labels = plt.xticks()
_ = plt.setp(labels, rotation=90)
```

```python
heatmap_seaborn = sns.heatmap(df_MAG_cellulosomes,cmap=mpl.cm.Blues, cbar_kws={'orientation':'horizontal'})
locs, labels = plt.xticks()
_ = plt.setp(labels, rotation=90)
```

```python
data_linearized_d = {}
for mag, row in df_MAG_abundance.iterrows():
    for i, val_col in enumerate(['101B', '102', '103', '104', '105', '106']):
        data_linearized_d[mag + "_{}".format(val_col)] = {"abundance": row[val_col],
                                                    "mag": mag,
                                                    "sample": val_col}
abundance_linearized = pd.DataFrame.from_dict(data_linearized_d, orient='index')
```

```python
top_ax1 = plt.subplot2grid((15,22), (0, 0), colspan=7)
top_ax2 = plt.subplot2grid((15,22), (0, 8), colspan=7)
top_ax3 = plt.subplot2grid((15,22), (0, 17), colspan=6)

ax1 = plt.subplot2grid((15,22), (1, 0), rowspan=14, colspan=16)
ax2 = plt.subplot2grid((15,22), (1, 16), rowspan=14, colspan=2)
ax3 = plt.subplot2grid((15,22), (1, 18), rowspan=14, colspan=4)

# A completely white colorscale
cmap1 = mpl.colors.ListedColormap(['w'])

# Previously used light green palette
cmap=sns.light_palette((0.19783576093349015, 0.6955516966063037, 0.3995301037444499), n_colors=100, as_cmap=True)

# Previously used light blue palette
sns.light_palette((0.22335772267769388, 0.6565792317435265, 0.8171355503265633), n_colors=10, as_cmap=True)




heatmap_cellulases = sns.heatmap(df_MAG_cellulases, vmax=15, annot=True, annot_kws = {'fontsize': 7},
                              mask=df_MAG_cellulases < 0.1,
                              cmap="Greens",
                              center=6.5,
                              cbar_kws={'orientation':'horizontal'}, 
                              cbar_ax=top_ax1, ax=ax1)

top_ax1.xaxis.tick_top()
top_ax1.set_xticklabels(["", 3, 6, 9, 12, ">15"])
top_ax1.set_aspect(0.05)
#heatmap_cellulases.ticklabel_format?
heatmap_cellulases.tick_params(axis='y', which='major', labelsize=8)

heatmap_cellulosomes = sns.heatmap(df_MAG_cellulosomes, vmax=10, annot=True, annot_kws = {'fontsize': 7},
                              mask=df_MAG_cellulosomes < 0.1,
                              cmap="Blues",
                              center=4.5,
                              cbar_kws={'orientation':'horizontal'}, 
                              cbar_ax=top_ax2, ax=ax2)

heatmap_cellulosomes = sns.heatmap(df_MAG_cellulosomes, vmax=10, annot=True, annot_kws = {'fontsize': 7},
                              mask=df_MAG_cellulosomes < 0.1,
                              cmap="Blues",
                              center=4.5,
                              cbar=False,
                              yticklabels=False, ax=ax2)

top_ax2.xaxis.tick_top()
top_ax2.set_xticklabels(["", 2, 4, 6, 8, ">10"])
top_ax2.set_aspect(0.05)

#labels = heatmap_cellulosomes.get_xticklabels()
#_ = plt.setp(labels, rotation=75)


#ax_splt = sns.barplot(x='sum', y=df_MAG_abundance.index, data=df_MAG_abundance, color=(0.9603888539940703, 0.3814317878772117, 0.8683117650835491), ax=ax3)
#ax_splt = sns.boxplot(x='abundance', y='mag', data=abundance_linearized, color='r', order=df_MAG_abundance.index, ax=ax3)

x = range(6)*99

y = []
for n in range(99):
    y += [n+0.5]*6
s = list((df_MAG_abundance_rev[['101B', '102', '103', '104', '105', '106']].values * 30.0).flatten())

yticks = [n+0.5 for n in range(99)]

ax3.scatter(x,y, s=s, c='r', alpha=0.9)
ax3.set_ylim(0,99)
ax3.set_yticks(yticks)
ax3.set_yticklabels(df_MAG_abundance_rev.index, fontsize=8)
ax3.yaxis.tick_right()

ax3.set_xlim(-0.5,5.5)
ax3.set_xticks(range(6))
ax3.set_xticklabels(['101B', '102', '103', '104', '105', '106'], rotation=90)

ax3.set_ylabel("")
ax3.xaxis.grid(False)
ax3.yaxis.grid(True)
ax3.set_xlabel("Average coverage\nper million reads")
ax3.xaxis.set_label_position('top')

ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.spines['bottom'].set_visible(False)
ax3.spines['left'].set_visible(False)

x = range(3)

y = [1]*3
real_s = [5.0, 2.0, 0.5]
annot = ["5x", "2x", "0.5x"]
s = [30.0*n for n in real_s]



top_ax3.scatter(x, y, s=s, c='r')

for i, txt in enumerate(annot):
    top_ax3.annotate(txt, (x[i]-0.2,y[i]+0.06))

top_ax3.spines['top'].set_visible(False)
top_ax3.spines['right'].set_visible(False)
top_ax3.spines['bottom'].set_visible(False)
top_ax3.spines['left'].set_visible(False)
top_ax3.xaxis.grid(False)
top_ax3.yaxis.grid(False)
top_ax3.set_yticks([])
top_ax3.set_xticks([])

#sns.despine()
#plt.tight_layout()

ax1.set_xlabel("A", fontsize=15)
ax1.xaxis.set_label_position('top')

ax2.set_xlabel("B", fontsize=15)
ax2.xaxis.set_label_position('top')

ax3.set_xlabel("C", fontsize=15)
ax3.xaxis.set_label_position('top')

plt.savefig("heatmaps_combined_circle_with_scale.pdf", dpi=1200)
```
